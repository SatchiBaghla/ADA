```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
```

Sharpening of image

```python
image = cv.imread('imageblr.jpg')
plt.imshow(image)
```

```python
#create the sharpening image
#kernel = np.ones((3,3), np.float32)
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
kernel1 = np.ones((5,5), np.float32)/30
#sharpen the image by applying filter2D() function
sharpened_image = cv.filter2D(src = image, ddepth=-1, kernel=kernel)
blur_image = cv.filter2D(image,-1, kernel1)

```

```python
kernel_sobel = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
sobel_image = cv.filter2D(image, -1, kernel_sobel)

ksobel_btm = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
sobelbtm_image = cv.filter2D(image, -1, ksobel_btm)

ksobel_lft = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
sobellft_image = cv.filter2D(image, -1, ksobel_lft)

ksobel_rt = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobelrt_image = cv.filter2D(image, -1, ksobel_rt)

kernel_custom = np.array([[1,0,0],[0,1,0],[0,0,1]])
custom_image = cv.filter2D(image, -1, kernel_custom)
```

```python
plt.figure(figsize=(14, 14))
#plot the sharpened image
plt.subplot(241), plt.title("Original"),plt.imshow(image)
plt.subplot(242), plt.title("Sharpening"),plt.imshow(sharpened_image)
plt.subplot(243), plt.title("Blurred"),plt.imshow(blur_image)
plt.subplot(244), plt.title("Sobel Top"),plt.imshow(sobel_image)
plt.subplot(345), plt.title("Sobel Bottom"),plt.imshow(sobelbtm_image)
plt.subplot(346), plt.title("Sobel Left"),plt.imshow(sobellft_image)
plt.subplot(347), plt.title("Sobel Right"),plt.imshow(sobelrt_image)
plt.subplot(348), plt.title("Custom Image"),plt.imshow(custom_image)
plt.show()
```

Noise in Image


```python
noisy_image = cv.imread('noisyimage.jpg')
plt.imshow(noisy_image)
```

```python
#apply noise removal techniques

#apply mean filtering
mean_filtered = cv.blur(noisy_image,(5,5))

#gaussian filtering
gaussian_blur = cv.GaussianBlur(noisy_image,(3,3),0)

#Median filter
median_blur = cv.medianBlur(noisy_image, 5)

#bilateral filter
bilateral_filter = cv.bilateralFilter(noisy_image, 9, 75, 75)

#apply non-local means filtering
nlm_filtered = cv.fastNlMeansDenoising(noisy_image, None, 30,7,21)


#
```
