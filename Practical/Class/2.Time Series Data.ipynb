```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

file_path = "D:/MCA/4th Trm/ADA/Practical/Class/NIFTY 50.csv"
df = pd.read_csv(file_path)

print(df.head())
df.info()
df.isnull().sum()
df.describe()
print(df.columns)
```

```python
# Initialize the figure
plt.figure(figsize=(20, 20))

plt.subplot(3, 2, 1)
sns.lineplot(data=df, x='Date', y='Open')
plt.title('Opening Price over Time')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.xticks(rotation=45)



```
