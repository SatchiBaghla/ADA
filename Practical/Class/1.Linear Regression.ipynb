```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv('D:/MCA/4th Trm/ADA/Practical/Class/Sleep_health_and_lifestyle_dataset.csv')
```

```python
print(data.isna().sum())
```

```python
feature_columns = 'Sleep Duration'
target_column = 'Stress Level'
x = data[[feature_columns]]
y = data[target_column]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.7, random_state=50)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('R^2 Score:', r2)
```

```python
print('Coefficient:', model.coef_[0])
print('Intercept:', model.intercept_)
```

```python


# Scatter plot for actual vs predicted
plt.scatter(X_test ,y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.title('Simple Linear Regression')
plt.xlabel('Sleep Duration')
plt.ylabel('Stress Level')
plt.legend()
plt.show()
```
