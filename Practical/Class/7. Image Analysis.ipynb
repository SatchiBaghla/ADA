```python
x = np.random.randint(0,5,(500,500))
img = cv.imread('imageclr1.jpg')
print("Image dimension: ", img.shape)
```

```python
img_resize = cv.resize(img, (10,10))
plt.subplot(321), plt.title("Original"), plt.imshow(img)
plt.subplot(322), plt.title("Resized Image"), plt.imshow(img_resize)
plt.subplot(323), plt.title("Red"), plt.imshow(img_resize[:,:,0])
plt.subplot(324), plt.title("Green"), plt.imshow(img_resize[:,:,1])
plt.subplot(325), plt.title("Blue"), plt.imshow(img_resize[:,:,2])
```

```python
img_resize2 = cv.resize(img, (1000,1000))
plt.subplot(321), plt.title("Original"), plt.imshow(img)
plt.subplot(322), plt.title("Resized Image"), plt.imshow(img_resize2)
plt.subplot(323), plt.title("Red"), plt.imshow(img_resize2[:,:,0])
plt.subplot(324), plt.title("Green"), plt.imshow(img_resize2[:,:,1])
plt.subplot(325), plt.title("Blue"), plt.imshow(img_resize2[:,:,2])
```
